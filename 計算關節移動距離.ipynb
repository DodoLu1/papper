```python
filename = '測試'
dataset = pd.read_csv('D:/openpose_data/揮手測試/'+filename+'/骨架資訊_all_標記.csv',header= None) 
dataset
```

```python
def Calculate_distance(x,y,z,x_1,y_1,z_1):
    if((x != 0) & (y != 0) & (z != 0) & (x_1 != 0) & (y_1 != 0) & (z_1 != 0)):
        d = (x-x_1)**2+(y-y_1)**2+(z-z_1)**2
        return d**0.5
    else:
        return 0
```
