```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

x = np.arange(0 , 10, 0.1)
y = softmax(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.title('Softmax Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))

x = np.linspace(-5, 5, 100)
y = softmax(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Softmax Function')
plt.show()
```
