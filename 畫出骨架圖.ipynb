```python
def skeleton_jpg(tt,na):
    output_folder = ('D:/openpose_data/揮手測試/'+name+'/骨架圖片/')
    # 將 list 轉換成 numpy array
    pose_array = np.array(tt)

    # 將每個關節點位的 (x, y, z) 分別抓出來
    x_coords = pose_array[0::3]
    y_coords = pose_array[1::3]
    z_coords = pose_array[2::3]

    # 將 (x, y, z) 合併成一個 tuple 並加入新的 list 中
    joint_coords = []
    for x, y, z in zip(x_coords, y_coords, z_coords):
        joint_coords.append((int(x), int(y), int(z)))

    # 將骨架資料轉換為NumPy陣列
    joint_coords = np.array(joint_coords)

    # 計算x、y、z軸的縮放比例
    scale_x = 0.3
    scale_y = 0.3
    scale_z = 0.3

    # 進行縮放操作
    joint_coords = joint_coords * np.array([scale_x, scale_y, scale_z])

    # 將縮放後的數值轉換為整數值
    joint_coords = joint_coords.astype(int)


    # 設定視窗大小
    win_width = 350
    win_height = 600

    # 建立視窗
    canvas = np.zeros((win_height, win_width, 3), dtype=np.uint8)

    # 定義節點顏色和線條顏色
    node_color = (255, 0, 0)
    line_color = (0, 255, 0)


    # 將骨架節點座標轉換為整數值
    #joint_coords = [(int(x), int(y), int(z)) for x, y, z in joint_coords]

    # 繪製骨架節點
    for node in joint_coords:
        cv2.circle(canvas, (node[0], node[1]), 5, node_color, -1)

    # 繪製骨架線條
    #for i in range(len(joint_coords) - 1):
    #    cv2.line(canvas, (joint_coords[i][0], joint_coords[i][1]), (joint_coords[i+1][0], joint_coords[i+1][1]), line_color, 2)
    for i in range(4):
        if joint_coords[i][0] != 0 and joint_coords[i][1] != 0 and joint_coords[i+1][0] != 0 and joint_coords[i+1][1] != 0:
            cv2.line(canvas, (joint_coords[i][0], joint_coords[i][1]), (joint_coords[i+1][0], joint_coords[i+1][1]), line_color, 2)
            
    if joint_coords[1][0] != 0 and joint_coords[1][1] != 0 and joint_coords[5][0] != 0 and joint_coords[5][1] != 0:           
        cv2.line(canvas, (joint_coords[1][0], joint_coords[1][1]), (joint_coords[5][0], joint_coords[5][1]), line_color, 2)

    for i in range(5,7):
        if joint_coords[i][0] != 0 and joint_coords[i][1] != 0 and joint_coords[i+1][0] != 0 and joint_coords[i+1][1] != 0:
            cv2.line(canvas, (joint_coords[i][0], joint_coords[i][1]), (joint_coords[i+1][0], joint_coords[i+1][1]), line_color, 2)

    if joint_coords[1][0] != 0 and joint_coords[1][1] != 0 and joint_coords[8][0] != 0 and joint_coords[8][1] != 0:         
        cv2.line(canvas, (joint_coords[1][0], joint_coords[1][1]), (joint_coords[8][0], joint_coords[8][1]), line_color, 2)

    if joint_coords[8][0] != 0 and joint_coords[8][1] != 0 and joint_coords[9][0] != 0 and joint_coords[9][1] != 0: 
        cv2.line(canvas, (joint_coords[8][0], joint_coords[8][1]), (joint_coords[9][0], joint_coords[9][1]), line_color, 2)

    for i in range(9,11):
        if joint_coords[i][0] != 0 and joint_coords[i][1] != 0 and joint_coords[i+1][0] != 0 and joint_coords[i+1][1] != 0:
            cv2.line(canvas, (joint_coords[i][0], joint_coords[i][1]), (joint_coords[i+1][0], joint_coords[i+1][1]), line_color, 2)

    if joint_coords[8][0] != 0 and joint_coords[8][1] != 0 and joint_coords[12][0] != 0 and joint_coords[12][1] != 0:        
        cv2.line(canvas, (joint_coords[8][0], joint_coords[8][1]), (joint_coords[12][0], joint_coords[12][1]), line_color, 2)

    for i in range(12,14):
        if joint_coords[i][0] != 0 and joint_coords[i][1] != 0 and joint_coords[i+1][0] != 0 and joint_coords[i+1][1] != 0:
            cv2.line(canvas, (joint_coords[i][0], joint_coords[i][1]), (joint_coords[i+1][0], joint_coords[i+1][1]), line_color, 2)
            
    #臉部點位線
    if joint_coords[0][0] != 0 and joint_coords[0][1] != 0 and joint_coords[15][0] != 0 and joint_coords[15][1] != 0:           
        cv2.line(canvas, (joint_coords[0][0], joint_coords[0][1]), (joint_coords[15][0], joint_coords[15][1]), line_color, 2)
    if joint_coords[0][0] != 0 and joint_coords[0][1] != 0 and joint_coords[16][0] != 0 and joint_coords[16][1] != 0:           
        cv2.line(canvas, (joint_coords[0][0], joint_coords[0][1]), (joint_coords[16][0], joint_coords[16][1]), line_color, 2)
    if joint_coords[15][0] != 0 and joint_coords[15][1] != 0 and joint_coords[17][0] != 0 and joint_coords[17][1] != 0:           
        cv2.line(canvas, (joint_coords[15][0], joint_coords[15][1]), (joint_coords[17][0], joint_coords[17][1]), line_color, 2)
    if joint_coords[16][0] != 0 and joint_coords[16][1] != 0 and joint_coords[18][0] != 0 and joint_coords[18][1] != 0:           
        cv2.line(canvas, (joint_coords[16][0], joint_coords[16][1]), (joint_coords[18][0], joint_coords[18][1]), line_color, 2)

    #腳掌點位線
    if joint_coords[11][0] != 0 and joint_coords[11][1] != 0 and joint_coords[24][0] != 0 and joint_coords[24][1] != 0:           
        cv2.line(canvas, (joint_coords[11][0], joint_coords[11][1]), (joint_coords[24][0], joint_coords[24][1]), line_color, 2)
    if joint_coords[11][0] != 0 and joint_coords[11][1] != 0 and joint_coords[22][0] != 0 and joint_coords[22][1] != 0:           
        cv2.line(canvas, (joint_coords[11][0], joint_coords[11][1]), (joint_coords[22][0], joint_coords[22][1]), line_color, 2)
    if joint_coords[22][0] != 0 and joint_coords[22][1] != 0 and joint_coords[23][0] != 0 and joint_coords[23][1] != 0:           
        cv2.line(canvas, (joint_coords[22][0], joint_coords[22][1]), (joint_coords[23][0], joint_coords[23][1]), line_color, 2)
        
    if joint_coords[14][0] != 0 and joint_coords[14][1] != 0 and joint_coords[21][0] != 0 and joint_coords[21][1] != 0:           
        cv2.line(canvas, (joint_coords[14][0], joint_coords[14][1]), (joint_coords[21][0], joint_coords[21][1]), line_color, 2)
    if joint_coords[14][0] != 0 and joint_coords[14][1] != 0 and joint_coords[19][0] != 0 and joint_coords[19][1] != 0:           
        cv2.line(canvas, (joint_coords[14][0], joint_coords[14][1]), (joint_coords[19][0], joint_coords[19][1]), line_color, 2)
    if joint_coords[19][0] != 0 and joint_coords[19][1] != 0 and joint_coords[20][0] != 0 and joint_coords[20][1] != 0:           
        cv2.line(canvas, (joint_coords[19][0], joint_coords[19][1]), (joint_coords[20][0], joint_coords[20][1]), line_color, 2)

    # 顯示圖片
    #cv2.imshow("Skeleton", canvas)
    cv2.imwrite(output_folder+str(na)+'.jpg',canvas)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
```
