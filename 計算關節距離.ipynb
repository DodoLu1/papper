```python
import pandas as pd
import numpy as np
```

```python
filename = '測試'
dataset = pd.read_csv('D:/openpose_data/揮手測試/'+filename+'/骨架資訊_all_標記.csv') 
dataset
```

```python
def Calculate_distance(x,y):
    if((x[0] != 0) & (x[1] != 0) & (x[2] != 0) ):
        d = (x[0]-y[0])**2+(x[1]-y[1])**2+(x[2]-y[2])**2
        return d**0.5
    else:
        return 0
```

```python
newlist = []
for i in range(len(dataset)):
    y_1 = dataset.iloc[i]
    y = [(y_1[0+1*3]+y_1[0+8*3])/2,(y_1[1+1*3]+y_1[1+8*3])/2,(y_1[2+1*3]+y_1[2+8*3])/2] #將脖子及髖關節點位相加除以2找出身體中心點位
    for j in range(25):
        x = dataset.iloc[i][0+j*3],dataset.iloc[i][1+j*3],dataset.iloc[i][2+j*3]
        newlist.append(Calculate_distance(x,y))
```

```python
sk_0 = np.reshape(newlist, (len(dataset), 25))
```

```python
tag_name = ['Nose', 'Neck', 'RShoulder', 'RElbow', 'RWrist', 'LShoulder', 'LElbow', 'LWrist', 'MidHip', 'RHip', 'RKnee', 'RAnkle', 'LHip', 'LKnee', 'LAnkle', 'REye', 'LEye', 'REar', 'LEar', 'LBigToe', 'LSmallToe', 'LHeel', 'RBigToe', 'RSmallToe', 'RHeel']
sk_data = pd.DataFrame(sk_0,columns=tag_name)
```
