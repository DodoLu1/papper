```python
dataset = pd.read_csv("D:/openpose_data/揮手測試/揮手測試_all_標記.csv") 
dataset
```

判斷手腕x軸是增加還是減少

```python
rl = []
for i in range(len(dataset['12'])-1):
    if dataset['12'][i] == 0 or dataset['12'][i+1] ==0:
        rl.append(0)
    else:
        if dataset['12'][i] == dataset['12'][i+1]:
            rl.append(0) 
        if dataset['12'][i] > dataset['12'][i+1]:
            rl.append(-1) 
        if dataset['12'][i] < dataset['12'][i+1]:
            rl.append(1) 
```

```python
tag_name = ['方向']
sk_data = pd.DataFrame(rl,columns=tag_name)
```

```python
sk_data.to_csv('D:/openpose_data/揮手測試/揮手測試_方向.csv',index=False)
```
